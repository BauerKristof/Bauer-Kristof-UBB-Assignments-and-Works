```python
import numpy as np
import matplotlib.pyplot as plt
import random

def felepitAdatok(tablazat):
    tablazatSora=len(tablazat)
    transzformaciosMatrix=[]
    eltolasiVektor=[]
    valoszinusegek=[]

    for i in range(tablazatSora):
        transzformaciosMatrix.append(np.array([[tablazat[i][0],tablazat[i][1]],[tablazat[i][2],tablazat[i][3]]]))
        eltolasiVektor.append(np.array([tablazat[i][4],tablazat[i][5]]))
        valoszinusegek.append(tablazat[i][6])
    return transzformaciosMatrix,eltolasiVektor,valoszinusegek,tablazatSora

def kirajzolFraktal(transzformaciosMatrix,eltolasiVektor,valoszinusegek,n):
    ax = plt.subplot()
    szin = ['blue','purple','yellow','brown','orange','red','green','lime']
    aktualisPont = np.array([0,0])
    
    for i in range(10000):
        rand = random.choices(range(n),valoszinusegek) 
        randValoszinuseg = rand[0]
        ujpont = np.matmul(transzformaciosMatrix[randValoszinuseg],aktualisPont)+eltolasiVektor[randValoszinuseg]
        ax.plot(ujpont[0],ujpont[1],color=szin[randValoszinuseg],marker='.')
        aktualisPont = ujpont
    plt.show()

serp_data = [[0.5, 0, 0, 0.5, 0, 0, 1/3],
    [0.5, 0, 0, 0.5, 0, 0.5, 1/3],
    [0.5, 0, 0, 0.5, 0.5, 0, 1/3]]

transzformaciosMatrix,eltolasiVektor,valoszinusegek,n=felepitAdatok(serp_data)
kirajzolFraktal(transzformaciosMatrix,eltolasiVektor,valoszinusegek,n)    
```

```python
def kirajzolMobiusFraktal(tablazat):
    tablazatSora = len(tablazat)
    
    valoszinusegek = []
    for i in range(tablazatSora):
        valoszinusegek.append(tablazat[i][4])

    ax = plt.subplot()
    z = complex(0,0)
    szin = ['blue','yellow','purple']

    for i in range(10000):
        rand = random.choices(range(tablazatSora),valoszinusegek)
        randValoszinuseg = rand[0]
        ujpont = (tablazat[randValoszinuseg][0] * z + tablazat[randValoszinuseg][1]) / (tablazat[randValoszinuseg][2] * z + tablazat[randValoszinuseg][3])
        
        valosPont=np.real(ujpont)
        imaginariusPont=np.imag(ujpont)
        ax.plot(valosPont,imaginariusPont,color=szin[randValoszinuseg],marker='.')
        z = ujpont
    plt.show();
        

ap_tri_data=[[1+np.sqrt(3),2+np.sqrt(3),-2-np.sqrt(3),5+3*np.sqrt(3),1/3],
    [(-1+1j*np.sqrt(3))*(1+np.sqrt(3)),(-1+1j*np.sqrt(3))*(2+np.sqrt(3)),-4-2*np.sqrt(3),10+6*np.sqrt(3),1/3],
    [(-1-1j*np.sqrt(3))*(1+np.sqrt(3)),(-1-1j*np.sqrt(3))*(2+np.sqrt(3)),-4-2*np.sqrt(3),10+6*np.sqrt(3),1/3]]

kirajzolMobiusFraktal(ap_tri_data)
```
